```python

df = pd.read_csv("C:/Users/user/Downloads/device_specs_structured_dataset.csv")


df["launch_date"] = (
    df["launch_date"]
    .astype(str)
    .str.replace(r"\s*\(Official\)", "", regex=True)
    .str.strip()
)
df["launch_date"] = pd.to_datetime(df["launch_date"], format="%B %d, %Y", errors="coerce")
df["launch_year"]  = df["launch_date"].dt.year.fillna(0).astype(int)


print(df["launch_year"].value_counts().sort_index())
print(f"Shape: {df.shape}")
```

```python

df["ram"]          = df["ram_raw"].str.extract(r"(\d+)").astype(float)
df["storage"]      = df["storage_raw"].str.extract(r"(\d+)").astype(float)
df["battery"]      = df["battery_raw"].str.extract(r"(\d+)").astype(float)
df["rear_camera"]  = df["rear_camera_raw"].str.extract(r"(\d+)").astype(float)
df["front_camera"] = df["front_camera_raw"].str.extract(r"(\d+)").astype(float)
df["fast_charging"]= df["fast_charging_raw"].str.extract(r"(\d+)").astype(float)


df.drop(["ram_raw","storage_raw","battery_raw",
         "rear_camera_raw","front_camera_raw",
         "fast_charging_raw"], axis=1, inplace=True)


for col in ["ram","storage","battery","rear_camera","list_price_inr"]:
    df[col].fillna(df[col].median(), inplace=True)
df["front_camera"].fillna(0, inplace=True)
df["fast_charging"].fillna(0, inplace=True)
for col in ["operating_system","chipset","display_type","sim_type","network_support"]:
    df[col].fillna(df[col].mode()[0], inplace=True)


bins   = [0, 10000, 20000, 35000, 60000, float("inf")]
labels = ["Budget", "Mid-Range", "Upper-Mid", "Premium", "Flagship"]
df["price_segment"] = pd.cut(df["price_inr"], bins=bins, labels=labels)


print(df.isnull().sum())
print(f"\n✅ Shape: {df.shape}")
```

```python
#PLOT 1 --Top 10 Brands by Number of Models
#----------------------------------------------
fig, ax = plt.subplots(figsize=(12, 6))
brand_counts=df ["brand"].value_counts().head(10)
bars=ax.barh(brand_counts.index[::-1],brand_counts.values[::-1],
            color=sns.color_palette("Blues_d",10))
for bar in bars:
    ax.text(bar.get_width()+0.1,bar.get_y()+bar.get_height()/2,
           str(int(bar.get_width())),va="center",fontsize=10)
ax.set_title("Top 10 Smartphone Brands by Number of Models",fontsize=14,fontweight="bold")
ax.set_xlabel("Number of Models")
ax.set_ylabel("Brand")
sns.despine()
plt.tight_layout()
plt.savefig("plot1_brand_count.png",bbox_inches="tight")
plt.show()
print("""
Observation:
All 10 brands have exactly 20 models each,indicating a well-balanced and evenly sampled dataset.This ensures
no single brand dominates the analysis,making brand comparisons fair and unbiased.
""")
```

```python
#PLOT 2 --Price Distribution by brand (Box Plot)
#-----------------------------------------------

fig,ax=plt.subplots(figsize=(14,7))
order = df.groupby("brand")["price_inr"].median().sort_values(ascending=False).index
sns.boxplot(data=df,x="brand",y="price_inr",order=order,
           palette="Set2",width=0.6,fliersize=4,ax=ax)
ax.set_title("Price Distribution Across Brands",fontsize=14,fontweight="bold")
ax.set_xlabel("Brand")
ax.set_ylabel("Price(INR)")
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_:f"₹{int(x):,}"))
ax.tick_params(axis="x",rotation=30)
sns.despine()
plt.tight_layout()
plt.savefig("plot2_price_by_brand.png",bbox_inches="tight")
plt.show()

print("""
Observation:
Apple and Samsung show the highest median prices,reflecting their
premium postioning.Brands like Realme and IQOO cluster in lower price ranges.Wide interquartile 
ranges in Oneplus and Google indicates they
target multiple market segments simultaneously.
""")
```

```python
# PLOT 3-- RAM vs Price (Scatter Plot)
#---------------------------------------

fig,ax=plt.subplots(figsize=(12,7))
sns.scatterplot(data=df,x="ram",y="price_inr",hue="brand",
               palette="tab10",alpha=0.75,s=90,
               edgecolor="white",linewidth=0.5,ax=ax)
ax.set_title("RAM vs Price -All Brands",fontsize=14,fontweight="bold")
ax.set_xlabel("RAM (GB)")
ax.set_ylabel("Price (INR)")
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_:f"₹{int(x):,}"))
ax.legend(title="Brand",bbox_to_anchor=(1.01,1),borderaxespad=0)
sns.despine()
plt.tight_layout()
plt.savefig("plot3_ram_vs_price.png",bbox_inches="tight")
plt.show()

print("""
Observation:
Ac clear positive correlation exists between RAM and price.Phones
with 12 GB+ RAM are predominantly in the premium and flagship segments.
Apple devices appear at lower RAM values yet command high prices,
highlighting that brand value outweighs raw specs for Apple.
""")
```

```python
#PLOT 4--Battery by Price Segment (Viloin Plot)

fig,ax=plt.subplots(figsize=(13,6))
segment_order=["Budget","Mid-Range","Upper-Mid","Premium","Flagship"]
sns.violinplot(data=df,x="price_segment",y="battery",
              order=segment_order,palette="pastel",
              inner="quartile",ax=ax)
ax.axhline(5000,color="tomato",linestyle="--",linewidth=1.2,label="5000 mAh reference")
ax.set_title("Battery Capacity by Price Segment",fontsize=14,fontweight="bold")
ax.set_xlabel("Price Segment")
ax.set_ylabel("Battery Capacity (mAh)")
ax.legend()
sns.despine()
plt.tight_layout()
plt.savefig("plot4_battery_by_segment.png",bbox_inches="tight")
plt.show()

print("""
Observation:
budget and Mid-Range phones consistently offer larger batteries
(4000-5000+ mAh) as a key selling point.Flagship phones sometimes 
have smaller batteries,trading capacity for slimmer designs and
premium build quality.
""")
```

```python
#PLOT 5 --Correlation Heatmap
fig,ax=plt.subplots(figsize=(10,8))
num_cols=["price_inr","ram","storage","battery",
         "rear_camera","front_camera","fast_charging","launch_year"]
corr=df[num_cols].corr()
mask=np.triu(np.ones_like(corr,dtype=bool))
sns.heatmap(corr,mask=mask,annot=True,fmt=".2f",cmap="coolwarm",
           center=0,linewidths=0.5,square=True,
           annot_kws={"size":10},ax=ax)
ax.set_title("Correlation Heatmap -Smartphone Specifications",fontsize=14,fontweight="bold")
plt.tight_layout()
plt.savefig("plot5_correlation_heatmap.png",bbox_inches="tight")
plt.show()

print("""
Observation:
Price shows the storngest positive correlation with RAM and storage.
Fast charging also correlates moderately with price,indicating it
is a premium feature.Battery capacity has a weak correaltion with
price,cofirming that larger batteries are common across all segments.
""")
```

```python
#PLOT 6--SmartPhones Launched Per Year
#-------------------------------------

fig,ax=plt.subplots(figsize=(11,5))
year_counts=df[df["launch_year"] !=0]["launch_year"].value_counts().sort_index()
ax.plot(year_counts.index,year_counts.values,marker="o",
       linewidth=2.5,color="steelblue",markersize=8,
       markerfacecolor="white",markeredgewidth=2)
for x,y,in zip(year_counts.index,year_counts.values):
    ax.text(x,y+1,str(y),ha="center",fontsize=10,fontweight="bold")
ax.set_title("Smartphones Launched Per Year",fontsize=14,fontweight="bold")
ax.set_xlabel("Year")
ax.set_ylabel("Number of Launches")
ax.set_xticks(year_counts.index)
sns.despine()
plt.tight_layout()
plt.savefig("plot6_launches_per_year.png",bbox_inches="tight")
plt.show()

print("""
Observation:
Smartphone launches have increased significantly from 2019 to 2025,
with 2025 seeing the highest number of releases.The sharp rise
reflects rapid market growth annd increasing competition among brands
to relaease multiple models annually.
""")
```
