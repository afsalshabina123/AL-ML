```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,14,12,18,20]
plt.plot(x,y,marker="o")
plt.title("simple line chart")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.grid("true")
plt.show()
```

```python
students=["A","B","C","D","E"]
marks=[85,70,90,65,95]
plt.bar(students,marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
students=["A","B","C","D","E"]
marks=[85,70,90,65,95]
plt.bar(students,marks,color=["black","blue","green","orange","Purple"])
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
activities=["Study","Exercise","Sleep","Hobbies"]
hours=[4,1,7,2]
plt.figure(figsize=(7,7))
plt.pie(hours,labels=activities,autopct="%1.1f%%",startangle=140)
plt.title("Daily Activities")
plt.tight_layout()
plt.show()
```
