```python
duplicate_count=df.duplicated().sum()
print(f"Number of duplicates:{duplicate_count}")
df.drop_duplicates(inplace=True)
```

```python
print(df.isnull().sum())
```

```python
missing=df.isnull().sum()
ax=missing.plot(kind="bar",color="skyblue")
plt.title("Missing values per column")
plt.ylabel("Count of Missing values")
plt.xlabel("Column")
plt.xticks(rotation=60)
for index,value in enumerate(missing):
    plt.text(index,value+5,str(value),ha="center",va="bottom",fontsize=10,fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
top_3_languages = df['belongs_to_collection'].value_counts().nlargest(3)

print("Top 3 Languages with Highest Number of Films:")
print(top_3_languages)
```

```python
import matplotlib.pyplot as plt

top_3_languages.plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=200,
    colors=["skyblue","lightcoral","lightgreen"],
    title="Top 3 Languages by Film Count"
)
plt.ylabel("TOP 3")
plt.show()
```

```python
top_languages = df['belongs_to_collection'].value_counts().head(3)
print(top_languages)
```

```python
df['genres'] = df['genres'].str.split('|')
df_exploded = df.explode('genres')
print(df_exploded[['title', 'genres']].head())
```

```python
df_exploded = df.assign(genres=df['genres'].str.split('|')).explode('genres')
genre_counts = df_exploded['genres'].value_counts()
print(genre_counts)
print(df.columns)
```

```python
genre_counts = df['genres'].str.split('|').explode().str.strip().value_counts()
plt.figure(figsize=(10, 6))
genre_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Movie Counts per Genre', fontsize=14)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
